# Make Dataset

```python
#!wget https://huggingface.co/datasets/Whispering-GPT/yannic-kilcher-transcript/resolve/main/data/train-00000-of-00001-af6113d3be1e7a86.parquet file1.parquet
#!wget https://huggingface.co/datasets/Whispering-GPT/lex-fridman-podcast/resolve/main/data/train-00000-of-00001-25f40520d4548308.parquet file2.parquet
#!wget https://huggingface.co/datasets/Whispering-GPT/whisper-transcripts-linustechtips/resolve/main/data/train-00000-of-00001-afb8a0941170b90c.parquet file3.parquet 
#!wget https://huggingface.co/datasets/Whispering-GPT/whisper-transcripts-ml-street-talk/resolve/main/data/train-00000-of-00001-c3e9273f83cdcb6d.parquet file4.parquet 
#!wget https://huggingface.co/datasets/Whispering-GPT/two-minute-papers/resolve/main/data/train-00000-of-00001-c4640e9210a7253c.parquet file5.parquet 
#!wget https://huggingface.co/datasets/Whispering-GPT/whisper-transcripts-the-verge/resolve/refs%2Fconvert%2Fparquet/Whispering-GPT--whisper-transcripts-the-verge/json-train.parquet file6.parquet 
import pandas as pd
df = pd.read_parquet('train-00000-of-00001-af6113d3be1e7a86.parquet')
df = df['text']
df.to_csv('training1.txt')
df = pd.read_parquet('train-00000-of-00001-25f40520d4548308.parquet')
df = df['text']
df.to_csv('training2.txt')
df = pd.read_parquet('train-00000-of-00001-c3e9273f83cdcb6d.parquet')
df = df['TRANSCRIPTION']
df.to_csv('training3.txt')
df = pd.read_parquet('train-00000-of-00001-c4640e9210a7253c.parquet')
df = df['TRANSCRIPTION']
df.to_csv('training4.txt')
df = pd.read_parquet('train-00000-of-00001-afb8a0941170b90c.parquet')
df = df['text']
df.to_csv('training5.txt')
df = pd.read_parquet('train-00000-of-00001-c4640e9210a7253c.parquet')
df = df['TRANSCRIPTION']
df.to_csv('training6.txt')
df = pd.read_parquet('json-train.parquet')
df = df['text']
df.to_csv('training7.txt')
```

```python
i = 1
for x in range(6):
  f = open("training" + str(i) +".txt", "r")
  ff = open("train.txt","a")
  fff = open("training" + str(i) + ".txt", "r")
  lines = fff.readlines()
  for Line in lines:
    s = f.readline()
    words = s.split()
    s = (" ".join(words[1:]))
    s = s.replace(".", ".\n")
    ff.write(s)
  i = i + 1
```

## Train

https://happytransformer.com/learning-parameters/

Add a file called "train.txt" before running the code below 



```python
from happytransformer import GENTrainArgs
args = GENTrainArgs(num_train_epochs=2)
happy_gen.train("train.txt", args=args)
```
